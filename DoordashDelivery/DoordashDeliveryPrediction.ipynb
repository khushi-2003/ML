```python
df = df[df['order_protocol'].notna()]
df = df[df['market_id'].notna()]
df = df[df['actual_delivery_time'].notna()]

                                         
```

```python
df.isna().sum()
```

```python
num_cols = df.select_dtypes(include=['float64']).columns
cat_cols = ['store_primary_category']
```

```python
for col in num_cols:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.isna().sum()
```

```python
df['duration'] = pd.to_datetime(df['actual_delivery_time'],format = '%H:%M:%S')- pd.to_datetime(df['created_at'],format = '%H:%M:%S')
```

```python
df['total_seconds'] = pd.to_timedelta(df['duration']).dt.total_seconds()
```

```python
#Separate X and Y features
X = df.drop(['total_seconds'], axis=1)
Y = df[['total_seconds']]
```
